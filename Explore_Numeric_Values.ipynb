```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Video_Games.csv')
data.head()
```

```python
data.shape
```

#Cleaning Categorical Data

```python
data = data.dropna()
len(data)
```

```python
data.info()
```

```python
data['User_Score'] = data['User_Score'].astype(float)
data['Year_of_Release'] = data['Year_of_Release'].astype(int)
data.info()
```

```python
data = data.drop(['Name', 'Platform', 'Genre', 'Publisher', 'Developer', 'Rating'], axis = 1)
data.head()
```

#Data Correlation

```python
sns.heatmap(data.corr(), annot = True, cmap = 'BuPu')
```

```python
sns.regplot(x = 'Global_Sales', y = 'NA_Sales', data = data)
```

```python
sns.regplot(x = 'Global_Sales', y = 'EU_Sales', data = data)
```

```python
sns.regplot(x = 'Global_Sales', y = 'JP_Sales', data = data)
```

```python
sns.regplot(x = 'Global_Sales', y = 'Other_Sales', data = data)
```

#Report Findings

Insights:
- Global Sales has high correlation above 80% with NA Sales, EU Sales, and Other Sales.
- NA Sales has high correlation above 70% with EU Sales and Other Sales.
